# Solutions to Last Assignment
This notebook shows the solutions to the Pauli eigenstate preparation assignment

```python
from qiskit import Aer, execute, QuantumCircuit
from qiskit.visualization import plot_bloch_multivector

sim = Aer.get_backend("statevector_simulator")

def display(qc):
    sv = execute(qc, sim).result().get_statevector()
    return plot_bloch_multivector(sv)
```

## Z eigenstates
The $z$ eigenstates are easy to prepare because the computation by default takes place in the computational basis. At the beginning of the algorithm the qubits are initialized to $|0\rangle$, so this is trivial. To prepare $|1\rangle$, we can flip the state with an $X$ gate.

```python
qc = QuantumCircuit(1)

display(qc)
```

```python
qc = QuantumCircuit(1)
qc.x(0)
```

```python
display(qc)
```

## X eigenstates
To prepare these eigenstates, a basis change from the $Z$ basis to the $X$ basis needs to be performed. Using $H|0\rangle = |+\rangle$, we can easily prepare $|+\rangle$. Then, we can either use $Z = HXH$ to find $ZH|0\rangle = |-\rangle$, or we can use $HX|0\rangle = H|1\rangle = |-\rangle$ to prepare the $X$ eigenstates.

```python
qc = QuantumCircuit(1)
qc.h(0)

display(qc)
```

```python
qc = QuantumCircuit(1)
qc.h(0)
qc.z(0)

display(qc)
```

## Y eigenstates
The phase gate $S$ allows us to prepare the $Y$ eigenstates. We have $S \ket{+} = \ket{r}$ and $S \ket{-} = \ket{l}$. Thus the $Y$ eigenstates can be prepared by adding a phase gate to the previous example.

```python
qc = QuantumCircuit(1)
qc.h(0)
qc.s(0)

display(qc)
```

```python
qc = QuantumCircuit(1)
qc.x(0)
qc.h(0)
qc.s(0)

display(qc)
```

```python
qc.draw()
```
